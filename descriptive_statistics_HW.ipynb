# Descriptive statistics problems

### Exercise 1

We will use Numpy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
import numpy as np
````

````python
# TODO

# 1) array of 100 elements with normal distribution.

array_1 = np.random.randn(100)

# 2) array of 100 elements with chi-square distribution, 3 degrees of freedom.

array_2 = np.random.chisquare(3, 100)

# 3) metrics

def describe_array(arr):
    print(f"Mean: {np.mean(arr)}")
    print(f"Median: {np.median(arr)}")
    print(f"Standard Dev: {np.std(arr)}")
    print(f"Variance: {np.var(arr)}")
    print(f"Min: {np.min(arr)}")
    print(f"Max: {np.max(arr)}")
    print(f"25th percentile {np.percentile(arr, 25)}")
    print(f"75th percentile: {np.percentile(arr, 75)}")
    iqr = np.percentile(arr, 75) - np.percentile(arr, 25)
    print(f"IQR: {iqr}")

print("Array 1 Statistics:")
describe_array(array_1)
print("-----")
print("Array 2 Statistics:")
describe_array(array_2)
````

````python
# TODO
# voy a asumir que esta data es un sample, y no el total de la poblacion. por ende el ddof para calcular sample std y no population std
data = [4, 2, 5, 8, 6]

def get_std_dev(list):
    return np.std(list, ddof=1)

result = get_std_dev(data)
result
````
